### Visualizing given points & initial centroids on graph

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x_coords = [2, 2, 11, 6, 6, 1, 5, 4, 10, 7, 9, 4, 3, 3, 6]
y_coords = [10, 6, 11, 9, 4, 2, 10, 9, 12, 5, 11, 6, 10, 8, 11]
centroids_x = [2, 5, 6]
centroids_y = [6, 10, 11]

plt.figure(figsize=(8, 6))
plt.grid(True)
plt.rcParams['axes.axisbelow'] = True
plt.scatter(x_coords, y_coords, s=75, edgecolor='k', label='Data Points')
plt.scatter(centroids_x, centroids_y, s=75, marker='X', edgecolor='k', c='r', label="Initial Centroids")
plt.title('Visualization of Given Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.xticks(range(0, 15))
plt.yticks(range(0, 15))
plt.legend()
plt.show()
```

### Defining functions necessary for K-Means

```python
cluster_0, cluster_1, cluster_2 = [], [], []

# Function to calculate the Euclidean distance
def euclidean_distance(p, q): # p and q are tuples
    return math.sqrt(sum((x - y)**2 for x, y in zip(p, q)))

# Function to assign clusters
def assign_clusters(X, centroids):
    clusters = []
    for x in X:
        distances = [euclidean_distance(x, centroid) for centroid in centroids]
        cluster = np.argmin(distances)
        clusters.append(cluster)
    return np.array(clusters)

# Function to recalculate centroids
def recalculate_centroids(X, clusters):
    new_centroids = [0, 0, 0]
    cluster_0.clear()
    cluster_1.clear()
    cluster_2.clear()
    for i in range(len(X)):
      if clusters[i] == 0:
        cluster_0.append(X[i])
      elif clusters[i] == 1:
        cluster_1.append(X[i])
      else:
        cluster_2.append(X[i])

    # print(cluster_0)
    # print(cluster_1)
    # print(cluster_2)
    #print(sum(x for x, y in cluster_0))
    new_centroids[0] = ((sum(x for x, y in cluster_0))/len(cluster_0), (sum(y for x, y in cluster_0))/len(cluster_0))
    new_centroids[1] = ((sum(x for x, y in cluster_1))/len(cluster_1), (sum(y for x, y in cluster_1))/len(cluster_1))
    new_centroids[2] = ((sum(x for x, y in cluster_2))/len(cluster_2), (sum(y for x, y in cluster_2))/len(cluster_2))
    return new_centroids

# Function to plot the clusters and centroids
def plot_clusters(X, centroids, clusters, iteration):
    plt.figure(figsize=(8, 6))
    cluster_0_x = [x for x, y in cluster_0]
    cluster_0_y = [y for x, y in cluster_0]
    cluster_1_x = [x for x, y in cluster_1]
    cluster_1_y = [y for x, y in cluster_1]
    cluster_2_x = [x for x, y in cluster_2]
    cluster_2_y = [y for x, y in cluster_2]
    plt.scatter(cluster_0_x, cluster_0_y, s=50, color='r', alpha=0.5, label=f'Cluster 0')
    plt.scatter(cluster_1_x, cluster_1_y, s=50, color='g', alpha=0.5, label=f'Cluster 1')
    plt.scatter(cluster_2_x, cluster_2_y, s=50, color='b', alpha=0.5, label=f'Cluster 2')
    plt.scatter(centroids[0][0], centroids[0][1], s=200, color='r', edgecolors='k', marker='o', label=f'Centroid 0')
    plt.scatter(centroids[1][0], centroids[1][1], s=200, color='g', edgecolors='k', marker='o', label=f'Centroid 1')
    plt.scatter(centroids[2][0], centroids[2][1], s=200, color='b', edgecolors='k', marker='o', label=f'Centroid 2')
    plt.title(f'Iteration {iteration}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    for point in centroids:
      plt.text(point[0], point[1], f'({point[0]}, {point[1]})', ha='center', va='bottom')
    plt.xticks(range(0, 15))
    plt.yticks(range(0, 15))
    plt.legend(loc=4)
    plt.grid(True)
    plt.show()
```

### 0th iteration of k-means

```python
X = [(x, y) for x, y in zip(x_coords, y_coords)]
centroids = [(x, y) for x, y in zip(centroids_x, centroids_y)]

clusters = assign_clusters(X, centroids)
new_centroids = recalculate_centroids(X, clusters)
plot_clusters(X, centroids, clusters, 0)
centroids = new_centroids
print("New Centroids after 0th iteration: " + str(centroids))
```

### 1st iteration of k-means

```python
clusters = assign_clusters(X, centroids)
new_centroids = recalculate_centroids(X, clusters)
plot_clusters(X, centroids, clusters, 1)
centroids = new_centroids
print("New Centroids after 1st iteration: " + str(centroids))
```
